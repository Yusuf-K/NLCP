### Теоретическая часть
![title](template.png)

Необходимо проанализировать схему для решения уравнения переноса $u_t + \lambda u_x = 0, \lambda = 1$ на вышеприведенном сеточном шаблоне с $\sigma = 0.2$

Все множество разностных схем:
$$u_m^{n+1} = \sum_{\mu, \nu}{\alpha_{\mu}^{\nu}(\tau, h)u_{m+\mu}^{n+\nu}}, \mu \not= 0, \nu\not=1$$
Для этого шаблона:
$$u_m^{n+1} = \alpha_{-1}^{-1}u_{m-1}^{n-1} + \alpha_{0}^{-1}u_{m}^{n-1} + \alpha_{0}^{0}u_{m}^{n} + \alpha_{-1}^{0}u_{m-1}^{n}$$ 

Запишем ошибку аппроксимации $\delta_{r} = L_{\tau,h}[u]^{\tau,h} - F_{\tau,h}$, где $L_{\tau,h}[u]^{\tau,h}$ - значение точного решения в узлах сетки, подставленное в разностный оператор, а $F_{\tau,h}$ - значение правой части в узлах сетки.

$\delta_{r} = L_{\tau,h}[u]^{\tau,h} - F_{\tau,h} \sim \alpha_{-1}^{-1} u(x_{m-1}, t^{n-1}) + \alpha_{0}^{-1}u(x_{m}, t^{n-1}) + \alpha_{0}^{0}u(x_m,t^n) + \alpha_{-1}^{0}u(x_{m-1},t^{n}) - u(x_m,t^{n+1}) $

Теперь разложим в ряд Тейлора относительно точки $(x_m,t^n):$
$$
\delta_{r} \sim  \alpha_{-1}^{-1}(u - h u_x - \tau u_t + \frac{h^2}{2}u_{xx} + \tau h u_{xt} + \frac{\tau^2}{2} u_{tt} -\frac{h^3}{6}u_{xxx} - \frac{h^2 \tau}{2}u_{xxt} - \frac{h \tau^2}{2}u_{xtt} - \frac{\tau^3}{6}u_{ttt} + O(\tau^4,h^4) ) + \alpha_{0}^{-1}(u -\tau u_{t} +\frac{\tau^2}{2} u_{tt} - \frac{\tau^3}{6}u_{ttt} + O(\tau^4))+
$$
$$
+ \alpha_{0}^{0} u + \alpha_{-1}^{0}(u - h u_x +\frac{h^2}{2}u_{xx} - \frac{h^3}{6} u_{xxx} + O(h^4)) - (u +\tau u_{t} +\frac{\tau^2}{2} u_{tt} + \frac{\tau^3}{6}u_{ttt} + O(\tau^4))
$$

Перепишем с учетом равенств $u_{tt}= -\lambda u_{tx} = \lambda^2 u_{xx},\ u_{ttt} = -\lambda u_{ttx} = \lambda^2 u_{txx} = -\lambda^3 u_{xxx},\ \sigma = \frac{\lambda \tau}{h}=\frac{\tau}{h}$:
$$
\delta_r \sim u(\alpha_{-1}^{-1}+\alpha_{0}^{-1}+\alpha_{-1}^{0}+\alpha_{0}^{0}-1) + u_{x}h(\alpha_{-1}^{-1}(\sigma-1)+\sigma\alpha_{0}^{-1}-\alpha_{-1}^{0}+\sigma) + u_{xx}\frac{h^2}{2}(\alpha_{-1}^{-1}(1-2\sigma+\sigma^2) + \sigma^2\alpha_{0}^{-1}+\alpha_{-1}^{0}-\sigma^2)+u_{xxx}\frac{h^3}{6}(\alpha_{-1}^{-1}(\sigma^3-3\sigma^2+3\sigma-1)+\sigma^3\alpha_{0}^{-1} - \alpha_{-1}^{0}+\sigma^3)
$$

При нашем $\lambda$ выполнено $\tau\sim h\Rightarrow O(\tau^4,h^4) = O(h^4)$

Чтобы размерности ошибок были одинаковыми надо разделить последнее уравнение на $\tau$

$$
\delta_0 = \alpha_{-1}^{-1}+\alpha_{0}^{-1}+\alpha_{-1}^{0}+\alpha_{0}^{0}-1\\
\delta_1 = \alpha_{-1}^{-1}(\sigma-1)+\sigma\alpha_{0}^{-1}-\alpha_{-1}^{0}+\sigma\\
\delta_2 = \alpha_{-1}^{-1}(1-2\sigma+\sigma^2) + \sigma^2\alpha_{0}^{-1}+\alpha_{-1}^{0}-\sigma^2\\
\delta_3 = \alpha_{-1}^{-1}(\sigma^3-3\sigma^2+3\sigma-1)+\sigma^3\alpha_{0}^{-1} - \alpha_{-1}^{0}+\sigma^3\Rightarrow\\
\delta_r = \underbrace{\frac{ \delta_0}{\tau}u}_{O(\frac{1}{h})} + \underbrace{\frac{h \delta_1}{\tau}u_x}_{O(1)} + \underbrace{\frac{h^2 \delta_2}{2\tau} u_{xx}}_{O(h)} + \underbrace{\frac{h^3 \delta_3}{6\tau} u_{xxx}}_{O(h^2)} +O(h^3) $$

Необходимо, чтобы $\delta_0 = 0$ (для аппроксимации разностной схемой уравнения)

Чтобы получить первый порядок аппроксимации ($O(h)$). Нужно дополнительно приравнять $\delta_1$ к нулю.

$\begin{equation*}
\begin{cases}
   \delta_0 = \alpha_{-1}^{-1}+\alpha_{0}^{-1}+\alpha_{-1}^{0}+\alpha_{0}^{0}-1 = 0
   \\
   \delta_1 = \alpha_{-1}^{-1}(\sigma-1)+\sigma\alpha_{0}^{-1}-\alpha_{-1}^{0}+\sigma = 0
 \end{cases} 
 \end{equation*}
$

Примем за независимые переменные $\alpha_{0}^{0}$ и $\alpha_{-1}^{0}$ и выразим оставшиеся коэффициенты через них и получим семейство схем первого порядка аппроксимации

$\begin{equation*}
\begin{cases}
   \alpha_{-1}^{-1} = -\frac{6}{5}\alpha_{-1}^{0}-\frac{1}{5}\alpha_{0}^{0}+\frac{2}{5}
   \\

 \end{cases} 
 \end{equation*}
$

Чтобы получить множество монотонных по Фридрихсу схем первого порядка аппроксимации надо решить систему неравенств относительно свободных переменных:

#### Практическая часть

Необходимо решить краевую задачу:
$\begin{equation*}
\begin{cases}
   u_t + \lambda u_x = 0, \lambda = 1 (t > 0, 0<x \le 2 ),
   \\
   u(0,x) = \varphi(x) (0<x \le 2),
   \\
   u(t,0) = 0 (0 < t \le 100 \tau).
 \end{cases} 
 \end{equation*}
 $

$\begin{equation*}
\varphi(x) = 
\begin{cases}
   10x-4, x \in [0.4,0.5],
   \\
   -10x+6, x \in [0.5,0.6],
   \\
   0, \text{иначе}.
 \end{cases} 
 \end{equation*}$ 

Вводим равномерную сетку на отрезке $[0,2]$ с числом узлов равным 201 ($h = 0.01$). Шаг по времени выбирается согласно введенному числу Куранта: $\sigma = \dfrac{\lambda \tau}{h} \Rightarrow \tau = \dfrac{\sigma h}{\lambda} = 0.002$.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Зададим шаг по времени, шаг по координате и их количество
num_coord_steps = 200
coord_step = 2.0 / num_coord_steps
time_step = 0.2 * coord_step
num_time_steps = 100

# Определим функцию phi(x)
def phi(x):
    triangle = 0.0
    if ((x >= 0.4) and (x <= 0.5)):
        triangle = 10*x-4
    elif ((x>=0.5) and (x<=0.6)):
        triangle = -10*x+6
    return triangle 
def point_to_scheme(x, y):
    a1 = -6.0/5*y-1.0/5*x+2.0/5
    a2 = 1.0/5*y-4.0/5*x+3.0/5
    a3 = x
    a4 = y
    return [a1,a2,a3,a4]
```

```python
# Зададим начальные условия
u = np.zeros(num_coord_steps)
for i in range(0, num_coord_steps):
    u[i] = phi(i*coord_step)
```

```python
x = np.linspace(0 , 2, num_coord_steps)
plt.plot(x, u)
```

```python
def calc(a1,a2,a3,a4,u,step):
    u_previous = np.zeros(num_coord_steps)
    u_current = np.zeros(num_coord_steps)
    u_current = u.copy()
    u_new = np.zeros(num_coord_steps)
    u_new[0] = 0 #boundary conditions
    for j in range (1, num_coord_steps):
        u_new[j] = 0.8*u_current[j] + 0.2*u_current[j-1]
    u_previous = u_current.copy()
    u_current = u_new.copy()
    #k = 0
    for i in range (1, step):
        #k += 1
        u_new[0] = 0 #boundary conditions
        #u_new[1] = 0.6*u_current[1] + 0.4*u_current[0]
        for j in range (1, num_coord_steps):
            u_new[j] = a1*u_previous[j-1] + a2*u_previous[j] + a3*u_current[j] + a4*u_current[j-1]
        u_previous = u_current.copy()   
        u_current = u_new.copy()
    return u_current    
```

```python
def print_calc(a1,a2,a3,a4,step):
    x_period = np.linspace(0,2,num_coord_steps)
    y_period = calc(a1,a2,a3,a4,u,step)
    plt.plot(x_period, y_period)
    x_exact = np.linspace(0,2,num_coord_steps)
    y_exact = np.zeros(num_coord_steps)
    for i in range(0,num_coord_steps):
        y_exact[i] = phi(x_exact[i] - step*time_step)
    plt.plot(x_exact, y_exact)
    
    plt.show()
```

##### 1п по четырем монотонным схемам первого порядка аппроксимации – вершинам области монотонных схем, включая схему с минимальной «аппроксимационной вязкостью» из (4т);

```python
# точка C
print_calc(1.0/4,0.0,3.0/4,0.0,num_time_steps/2)
print_calc(1.0/4,0.0,3.0/4,0.0,num_time_steps)
```

```python
# точка A
print_calc(0.0,2.0/3,0.0,1.0/3,num_time_steps/2)
print_calc(0.0,2.0/3,0.0,1.0/3,num_time_steps)
```

```python
# точка B
print_calc(2.0/5,3.0/5,0.0,0.0,num_time_steps/2)
print_calc(2.0/5,3.0/5,0.0,0.0,num_time_steps)
```

```python
# точка D
print_calc(0.0,0.0,4.0/5,1.0/5,num_time_steps/2)
print_calc(0.0,0.0,4.0/5,1.0/5,num_time_steps)
```

```python
#точка G
coeffs = point_to_scheme(0.98,-0.35)
print_calc(coeffs[0], coeffs[1], coeffs[2], coeffs[3],num_time_steps/2)
print_calc(coeffs[0], coeffs[1], coeffs[2], coeffs[3],num_time_steps)
```

```python
coeffs = point_to_scheme(1.14,-0.24)
print_calc(coeffs[0], coeffs[1], coeffs[2], coeffs[3],num_time_steps/2)
print_calc(coeffs[0], coeffs[1], coeffs[2], coeffs[3],num_time_steps)
```

```python
print_calc(0.4, -0.4, 1.2, -0.2,num_time_steps/2)
print_calc(0.4, -0.4, 1.2, -0.2,num_time_steps)
```

```python
def calc_hybrid(a1,a2,a3,a4,b1,b2,b3,b4,u,step):
    u_previous = np.zeros(num_coord_steps)
    u_current = np.zeros(num_coord_steps)
    u_current = u.copy()
    u_new = np.zeros(num_coord_steps)
    u_new[0] = 0 #boundary conditions
    for j in range (1, num_coord_steps):
        u_new[j] = 0.8*u_current[j] + 0.2*u_current[j-1]
    u_previous = u_current.copy()
    u_current = u_new.copy()
    for i in range (1, step):
        u_new[0] = 0 #boundary conditions
#         u_new[1] = 0.6*u_current[j] + 0.4*u_current[j-1]
        for j in range (1, num_coord_steps):
            calc = a1*u_previous[j-1] + a2*u_previous[j] + a3*u_current[j] + a4*u_current[j-1]
            if ((min(u_current[j-1],u_previous[j-1]) <= calc) and (max(u_current[j-1],u_previous[j-1]) >= calc)):
                u_new[j] = calc
            else:
                u_new[j] = b1*u_previous[j-1] + b2*u_previous[j] + b3*u_current[j] + b4*u_current[j-1]

#         u_new[num_space_steps-1] =  0.6*u_current[num_space_steps-1] + 0.4*u_current[num_space_steps-2]
        u_previous = u_current.copy()   
        u_current = u_new.copy()
    return u_current    
```

```python
def print_calc_hybrid(a1,a2,a3,a4,b1,b2,b3,b4,step):
    x_period = np.linspace(0,2,num_coord_steps)
    y_period = calc_hybrid(a1,a2,a3,a4,b1,b2,b3,b4,u,step)
    plt.plot(x_period, y_period)
    x_exact = np.linspace(0,2,num_coord_steps)
    y_exact = np.zeros(num_coord_steps)
    for i in range(0,num_coord_steps):
        y_exact[i] = phi(x_exact[i] - step*time_step)
    plt.plot(x_exact, y_exact)
    
    plt.show()
```

```python
coeffs1 = point_to_scheme(1.14,-0.24)
coeffs2 = point_to_scheme(1.26,-0.16)
print_calc_hybrid(coeffs1[0], coeffs1[1], coeffs1[2], coeffs1[3],coeffs2[0], coeffs2[1], coeffs2[2], coeffs2[3], num_time_steps/2)
print_calc_hybrid(coeffs1[0], coeffs1[1], coeffs1[2], coeffs1[3],coeffs2[0], coeffs2[1], coeffs2[2], coeffs2[3], num_time_steps)
```

```python
def calc_hybrid_3(a1,a2,a3,a4,b1,b2,b3,b4,c1,c2,c3,c4,u,step):
    u_previous = np.zeros(num_coord_steps)
    u_current = np.zeros(num_coord_steps)
    u_current = u.copy()
    u_new = np.zeros(num_coord_steps)
    u_new[0] = 0 #boundary conditions
    for j in range (1, num_coord_steps):
        u_new[j] = 0.8*u_current[j] + 0.2*u_current[j-1]
    u_previous = u_current.copy()
    u_current = u_new.copy()
    for i in range (1, step):
        u_new[0] = 0 #boundary conditions
#         u_new[1] = 0.6*u_current[j] + 0.4*u_current[j-1]
        for j in range (1, num_coord_steps):
            calc1 = a1*u_previous[j-1] + a2*u_previous[j] + a3*u_current[j] + a4*u_current[j-1]
            if ((min(u_current[j-1],u_previous[j-1]) <= calc1) and (max(u_current[j-1],u_previous[j-1]) >= calc1)):
                u_new[j] = calc1
            else:
                calc2 = b1*u_previous[j-1] + b2*u_previous[j] + b3*u_current[j] + b4*u_current[j-1]
                if ((min(u_current[j-1],u_previous[j-1]) <= calc2) and (max(u_current[j-1],u_previous[j-1]) >= calc2)):
                    u_new[j] = calc2
                else:
                    u_new[j] = c1*u_previous[j-1] + c2*u_previous[j] + c3*u_current[j] + c4*u_current[j-1]
#         u_new[num_space_steps-1] =  0.6*u_current[num_space_steps-1] + 0.4*u_current[num_space_steps-2]
        u_previous = u_current.copy()   
        u_current = u_new.copy()
    return u_current    
```

```python
def print_calc_hybrid_3(a1,a2,a3,a4,b1,b2,b3,b4,c1,c2,c3,c4,step):
    x_period = np.linspace(0,2,num_coord_steps)
    y_period = calc_hybrid_3(a1,a2,a3,a4,b1,b2,b3,b4,c1,c2,c3,c4,u,step)
    plt.plot(x_period, y_period)
    x_exact = np.linspace(0,2,num_coord_steps)
    y_exact = np.zeros(num_coord_steps)
    for i in range(0,num_coord_steps):
        y_exact[i] = phi(x_exact[i] - step*time_step)
    plt.plot(x_exact, y_exact)
    
    plt.show()
```
